```python

import pandas as pd
```

```python
dtypes = {
    
}

masses = pd.read_table(
    'McConnachie_masses.txt',
    delimiter='\t',
    skip_blank_lines=True,
    skiprows=1
).drop(range(0, 2))

masses
```

```python
sizes = pd.read_table(
    'McConnachie_sizes.txt',
    delimiter='\t',
    skiprows=5,
    skip_blank_lines=True,
).drop(range(0, 2))

sizes
```

```python
sizes[masses['Galaxy'] != sizes['Galaxy']]
```

```python
masses[masses['Galaxy'] != sizes['Galaxy']]
```

```python
data:pd.DataFrame = sizes.loc[:, ('Galaxy', 'r_h.1')].merge(masses.loc[:, ('Galaxy', 'M_sstarf')], how='outer', copy=True, left_index=True, right_index=True)
```

```python
dat = data.copy()

dat['M_sstarf'] = data['M_sstarf'].apply(pd.to_numeric, errors='coerce')

dat['M_stars'] = dat['M_sstarf'] * 1e6 # solar masses

dat['r_h.1'] = data['r_h.1'].apply(pd.to_numeric, errors='coerce')

dat.dropna(inplace=True)
dat.dtypes
```

```python
dat.to_pickle('mcconnachie_data.pickle')
```
